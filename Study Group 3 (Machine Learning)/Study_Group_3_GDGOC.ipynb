# Discover and Visualize the Data to Gain Insights

```python
sns.pairplot(housing, corner=True, diag_kind="kde", plot_kws={'alpha': 0.7})
plt.suptitle("Scatter Plot Semua Negara", y=1.02)
plt.show()
```

```python
housing.plot(kind="scatter", x="Austria", y="Year")
```

```python
plt.figure(figsize=(12, 8))
for country in housing.columns[1:]:
    plt.scatter(housing['Year'], housing[country], label=country, alpha=0.7)

plt.xlabel("Year")
plt.ylabel("Value")
plt.legend(title="Negara", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.show()
```

# Looking for Correlations

```python
corr_matrix = housing.select_dtypes(include=['number']).corr()
```

```python
corr_matrix["Year"].sort_values(ascending=False)
```
